```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# Importing dataset
epl_df = pd.read_csv('./EPL_20_21.csv')
```

```python
# Columns in the dataset
epl_df.columns
```

```python
# Number of players in each club
epl_df['Club'].value_counts()
```

```python
# Players Position Wise
epl_df['Position'].value_counts()
```

```python
# Creating variables Minutes played per match and Goals Scored per match
epl_df['Min_Per_Match'] = epl_df['Mins']/epl_df['Matches']
epl_df['Goals_Per_Match'] = epl_df['Goals']/epl_df['Matches']
epl_df.head()
```

```python
# Goals scored in the season
Total_Goals =  epl_df['Goals'].sum()
print(Total_Goals)
```

```python
# Penalty goals and Penalty Attempts in the season
Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print("Penalties Scored:", Total_Penalty_Goals)

Total_Penalty_Attempted = epl_df['Penalty_Attempted'].sum()
print("Penalties Attempted:", Total_Penalty_Attempted)
```

```python
# Pie Chart for Penalties
plt.figure(figsize=(10,8))
Penalties_Not_Scored = Total_Penalty_Attempted - Total_Penalty_Goals
data = [Penalties_Not_Scored, Total_Penalty_Goals]
labels = ['Penalties Not Scored', 'Penalties Scored']
color = sns.color_palette("tab10")
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# FW Players
epl_df[epl_df['Position'] == 'FW']
```

```python
# Nations of the players
np.size(epl_df['Nationality'].unique())
```

```python
# Plots showing nationality distribution of players
nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind='bar', figsize = (10, 8), color = sns.color_palette("flare"))
```

```python
# Top 10 Clubs with Largest Number of squad members
epl_df['Club'].value_counts().nlargest(10).plot(kind='bar', color = sns.color_palette("magma"))
```

```python
# Clubs with Least number of squad members
epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color = sns.color_palette("magma"))
```

```python
# Players based on age group

AgeUnder_20 = epl_df[epl_df['Age'] <= 20]
Age_20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
Age_25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
Above_30 = epl_df[epl_df['Age'] > 30]

x = np.array([AgeUnder_20['Name'].count(), Age_20_25['Name'].count(), Age_25_30['Name'].count(), Above_30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players Age Wise', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

```python
# Total under 20 player club wise
Players_Under_20 = epl_df[epl_df['Age'] < 20]
Players_Under_20['Club'].value_counts().plot(kind='bar', color = sns.color_palette("magma"))
```

```python
# Under-20 Players in Chelsea
Players_Under_20[Players_Under_20['Club'] == 'Chelsea']
```

```python
# Under-20 Players in Manchester City
Players_Under_20[Players_Under_20['Club'] == 'Manchester City']
```

```python
# Average player age club-wise
plt.figure(figsize=(10,8))
sns.boxplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
```

```python
# Average player age club-wise
num_player = epl_df.groupby('Club').size()
data  = (epl_df.groupby('Club')['Age'].sum())/ num_player
data.sort_values(ascending = False)
```

```python
# Bar plot for average age
data.plot(kind='bar', color = sns.color_palette("magma"))
```

```python
# Club wise Goals Scored
Club_Goals = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style = "whitegrid", color_codes = True)
gc = sns.barplot(x = 'Club', y = 'Goals', data = Club_Goals.sort_values(by="Goals"), palette= 'Set2')
gc.set_xlabel("Club", fontsize = 15)
gc.set_ylabel("Goals", fontsize = 20)
plt.title('Club vs Goals', fontsize = 20)
plt.xticks(rotation = 90)
plt.rcParams['figure.figsize'] = (10, 8)
```

```python
# Club wise Assists
Club_Assist = pd.DataFrame(epl_df.groupby('Club')['Assists'].sum())
sns.set_theme(style = 'darkgrid', color_codes=True)
ac = Club_Assist.plot(kind='bar', color = sns.color_palette("Set2"))
ac.set_xlabel('Clubs', fontsize = 15)
ac.set_ylabel('Assists', fontsize = 15)
plt.rcParams['figure.figsize'] = (10, 8)
plt.xticks(rotation=90)
plt.title('Club vs Assists', fontsize = 20)
```

```python
# Top 5 Players with most 
top_Goals = epl_df[['Name', 'Club', 'Goals']].sort_values(by='Goals', ascending = False).head(5)
top_Goals
```

```python
# Top 5 Players with most Assists
top_Assists = epl_df[['Name', 'Club', 'Matches','Assists']].nlargest(n=5, columns = 'Assists')
top_Assists
```

```python
# Players with most number of Matches played
most_Matches_Played = epl_df[['Name', 'Club', 'Matches']].nlargest(n=5, columns = 'Matches')
most_Matches_Played
```

```python
top_Goals_Per_Match = epl_df[['Name', 'Club', 'Goals','Matches','Goals_Per_Match']].nlargest(n=5, columns = 'Goals_Per_Match')
top_Goals_Per_Match
```

```python
# Assisted Goals vs Individual Goals
assists = epl_df['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Individual Goals', 'Assisted Goals']
color = sns.color_palette('Set1')
plt.figure(figsize=(10,8))
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# Players with most number of Yellow Cards
yellow_Cards = epl_df[['Name', 'Club', 'Yellow_Cards']].nlargest(n=5, columns = 'Yellow_Cards')
yellow_Cards
```

```python
# Yellow Card Plot
yellow_Card = epl_df.sort_values(by='Yellow_Cards', ascending = False)[:10]
yc = sns.barplot(x=yellow_Card['Name'], y=yellow_Card['Yellow_Cards'], label = 'Players', color='blue')
yc.set_xticklabels(yc.get_xticklabels(), rotation=70)
```

```python
# Players with most number of Red Cards
red_Cards = epl_df[['Name', 'Club', 'Red_Cards']].nlargest(n=5, columns = 'Red_Cards')
red_Cards
```

```python
# Red Card Plot
red_Card = epl_df.sort_values(by='Red_Cards', ascending = False)[:10]
rc = sns.barplot(x=red_Card['Name'], y=red_Card['Red_Cards'], label = 'Players', color='cyan')
rc.set_xticklabels(rc.get_xticklabels(), rotation=70)
```
